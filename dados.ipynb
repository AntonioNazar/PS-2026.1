```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
path_df_balanco_dia ="dataset_principal/balanco_servico_dia.csv"

df_balanco_dia = pd.read_csv(path_df_balanco_dia)

df_balanco_dia.head()
```

```python
df_balanco_dia_tratado = df_balanco_dia.drop(columns=['km_apurada','versao','datetime_ultima_atualizacao','id_execucao_dbt'])
df_balanco_dia_tratado.head()
```

```python
path_df_servico_historico = "dataset_principal/sumario_servico_dia_historico.csv"

df_servico_historico = pd.read_csv(path_df_servico_historico)

df_servico_historico.head()
```

```python
df_servico_historico[['bairro_inicio', 'bairro_fim']] = df_servico_historico['vista'].str.split('-', expand=True)
df_servico_historico_tratado = df_servico_historico.drop(columns=['datetime_ultima_atualizacao','valor_subsidio_pago','vista'])
```

```python
df_final = df_servico_historico_tratado.merge(df_balanco_dia_tratado, on=['servico', 'data','consorcio'], how='inner')

df_final['bairro_inicio'] = df_final['bairro_inicio'].str.strip().str.upper()

df_final['bairro_fim'] = df_final['bairro_fim'].str.strip().str.upper()
```

```python
df_final.head()
```

```python
col = df_final.pop('bairro_inicio')
df_final.insert(4, 'bairro_inicio', col)

col = df_final.pop('bairro_fim')
df_final.insert(5, 'bairro_fim', col)
```

```python
df_final = df_final.dropna().reset_index(drop=True)
```

```python
path_df_dados_pluviometricos = "dados_pluviometricos_extraidos/dados_pluviometricos_consolidados.csv"

df_dados_pluviometricos = pd.read_csv(path_df_dados_pluviometricos)

df_dados_pluviometricos['Bairro'] = df_dados_pluviometricos['Bairro'].str.strip().str.upper()


df_dados_pluviometricos.head(20)
```

```python
df_dados_pluviometricos = df_dados_pluviometricos.rename(columns={'Dia':'data'})
```

```python

df_reduzido = df_dados_pluviometricos[df_dados_pluviometricos['Hora'].astype(str) == '00:00:00'].copy()

df_reduzido['data'] = pd.to_datetime(df_reduzido['data']) - pd.Timedelta(days=1)

df_reduzido['data'] = df_reduzido['data'].astype(str)

df_reduzido = df_reduzido.drop(columns=['Hora'])

unique_bairros = df_reduzido['Bairro'].unique()

df_final = df_final[df_final['bairro_inicio'].isin(unique_bairros) | df_final['bairro_fim'].isin(unique_bairros)]
```

```python
df_cruzado = df_final.merge(

    df_reduzido, 

    left_on=['data', 'bairro_inicio'], 

    right_on=['data', 'Bairro'],       

    how='left'                        

)
df_cruzado = df_cruzado.merge(
    df_reduzido, 

    left_on=['data', 'bairro_fim'], 

    right_on=['data', 'Bairro'],       

    how='left'                        

)
df_cruzado = df_cruzado.drop(columns=['Bairro_x','Bairro_y'])
df_cruzado = df_cruzado.rename(columns={'24 h_x':'chuvas_no_bairro_inicio'})
df_cruzado = df_cruzado.rename(columns={'24 h_y':'chuvas_no_bairro_fim'})
df_cruzado.head()
```
